### Remove the missing value from the Dataset

```python
Data.dropna(inplace = True)
```

### Check the duplicate values

```python
Duplicate_values = Data.duplicated()
```

```python
print(Duplicate_values)
```

### Remove the Duplicate Values from the Data

```python
Data.drop_duplicates(inplace = True)
```

```python
Duplicate_values = Data.duplicated()
print(Duplicate_values)
```

### Now perform the statistics function on it

```python
Data.info
```

```python
Data.describe()
```

```python
sns.histplot(Data['Area'])
plt.title("Number of unemployment rate based on Area")
plt.show()
```

```python
sns.histplot(y = Data['Region'])
plt.title("Number of unemployment rate based on Area")
plt.show()
```

```python
Data[' Date'] = pd.to_datetime(Data[' Date'], dayfirst=True)
```

### Remove the extra space from the every columns

```python
Data.columns = Data.columns.str.strip()
```

### Unemployment Rate Trend Over Time

```python
plt.figure(figsize =(12,6))
sns.lineplot(x = Data['Date'],y =Data['Estimated Unemployment Rate (%)'],data = Data)
plt.title("Unemployment Rate Trend Over Time")
plt.show()
```

### Area-wise Unemployment Comparison

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=Data, x='Area', y='Estimated Unemployment Rate (%)')
plt.title("Unemployment Rate by Area Type")
plt.show()
```

### Region-wise Average Unemployment

```python
plt.figure(figsize=(12,6))
sns.barplot(data=Data, x='Region', y='Estimated Unemployment Rate (%)', estimator='mean')
plt.xticks(rotation=90)
plt.title("Average Unemployment Rate by Region")
plt.show()
```
